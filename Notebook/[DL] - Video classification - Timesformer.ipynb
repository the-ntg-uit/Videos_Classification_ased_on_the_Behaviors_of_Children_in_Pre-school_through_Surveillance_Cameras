# Train Timesformer

```python
from torch.optim import AdamW
from video_transformers import VideoModel
from video_transformers.backbones.transformers import TransformersBackbone
from video_transformers.data import VideoDataModule
from video_transformers.heads import LinearHead
from video_transformers.trainer import trainer_factory
from video_transformers.utils.file import download_ucf6


import torch 
backbone = TransformersBackbone("facebook/timesformer-base-finetuned-k400", num_unfrozen_stages=1)

datamodule = VideoDataModule(
    train_root="/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Data train model 2 label/Folder/train",
    val_root="/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Data train model 2 label/Folder/val",
    batch_size=12,
    num_workers=4,
    num_timesteps=8,
    preprocess_input_size=224,
    preprocess_clip_duration=1,
    preprocess_means=backbone.mean,
    preprocess_stds=backbone.std,
    preprocess_min_short_side=256,
    preprocess_max_short_side=320,
    preprocess_horizontal_flip_p=0.5,
)

head = LinearHead(hidden_size=backbone.num_features, num_classes=datamodule.num_classes)
model = VideoModel(backbone, head)

optimizer = AdamW(model.parameters(), lr=1e-4)

Trainer = trainer_factory("single_label_classification")
trainer = Trainer(datamodule, model, optimizer=optimizer, max_epochs=3)

trainer.fit()
```

```python
trainer.save_checkpoint("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[3] Model/Timesformer")
```

```python
import predict
from os import listdir
from os.path import isfile, join
import pandas as pd
from IPython.display import clear_output
```

```python
new_model.task = "single_label_classification"
```

```python
def evaluate(model, abn_path, n_path):
  abn_file = [f for f in listdir(abn_path) if isfile(join(abn_path, f))]
  n_file = [f for f in listdir(n_path) if isfile(join(n_path, f))]    
  df = pd.DataFrame({"path" : [], "true label" : [], "predict" : []})

  for i in abn_file:
    df.loc[-1] = {"path" : str(abn_path + '/' + i), "true label" : str(1), "predict" : None}
    df.index = df.index + 1  # shifting index
    df = df.sort_index()  # sorting by index

  for i in n_file:
    df.loc[-1] = {"path" : str(n_path + '/' + i), "true label" : str(0), "predict" : None}
    df.index = df.index + 1  # shifting index
    df = df.sort_index()  # sorting by index

  for i in range(len(df)):
    result = predict.predict(model=new_model, 
                    video_or_folder_path= df.at[i, "path"], 
                    preprocessor_config=  {"means": [0.45, 0.45, 0.45], "stds": [0.225, 0.225, 0.225], "min_short_side": 256, "input_size": 224, "clip_duration": 1, "num_timesteps": 8}, 
                    labels=["Abnormal", "Normal"])
    if result["predictions"]["Abnormal"] > result["predictions"]["Normal"]:
      df.at[i, "predict"] = str(1)
    else:
      df.at[i, "predict"] = str(0)
    clear_output(wait=True)
    print(i, '/', len(abn_file+n_file))

  return df
```

```python
abn_path = "/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Data train model 2 label/Folder/test/Abnormal"
n_path = "/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Data train model 2 label/Folder/test/Normal"
df = evaluate(new_model, abn_path, n_path)
```

```python
from sklearn.metrics import classification_report
print(classification_report(df["true label"], df["predict"], target_names=["Normal", "Abnormal"], digits = 4))
```

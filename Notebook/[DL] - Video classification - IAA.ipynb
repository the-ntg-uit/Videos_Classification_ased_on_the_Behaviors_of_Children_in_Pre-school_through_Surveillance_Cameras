```python
import pandas as pd
import numpy as np
```

```python
PT=pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Độ đồng thuận/DATA_DONGTHUAN_PT.xlsx")
DN=pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Độ đồng thuận/DATA_DONGTHUAN_DN.xlsx")
GT=pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Độ đồng thuận/DATA_DONGTHUAN_GT.xlsx")
HT=pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Độ đồng thuận/DATA_DONGTHUAN_HuynhT.xlsx")
MT=pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/Độ đồng thuận/DATA_DONGTHUAN_MinhTam.xlsx")
```

```python
PT['label']=PT['label'].replace(8,6)
HT['label']=HT['label'].astype(int)
GT['label']=GT['label'].astype(int)
DN['label']=DN['label'].astype(int)
```

```python
label_PT=PT['label'].replace([0,1,2,3,4,6],1).replace([7,5],0)

label_DN=DN['label'].replace([0,1,2,3,4,6],1).replace([7,5],0)

label_GT=GT['label'].replace([0,1,2,3,4,6],1).replace([7,5],0)

label_HT=HT['label'].replace([0,1,2,3,4,6],1).replace([7,5],0)

label_MT=MT['label'].replace([0,1,2,3,4,6],1).replace([7,5],0)
```

```python
lst=[label_PT.values,label_DN.values,label_GT.values,label_HT.values,label_MT.values]
name=['PT','DN','GT','HT','MT']
cohen_kappa=[]
```

```python
from sklearn.metrics import cohen_kappa_score
for i in range(len(lst)):
  for j in range(i+1,len(name)):
    print("độ đồng thuận giữa ",name[i], "và ", name[j], "là: ", cohen_kappa_score(lst[i],lst[j]))
    cohen_kappa.append(cohen_kappa_score(lst[i],lst[j]))
print("mean: ", np.mean(cohen_kappa))
```

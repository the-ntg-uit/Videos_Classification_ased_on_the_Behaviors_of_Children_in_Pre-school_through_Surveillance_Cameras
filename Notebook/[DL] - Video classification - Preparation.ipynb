```python
def splitter(A):
    B = A[0:len(A)//3]
    C = A[len(A)//3:(len(A)//3*2)]
    D = A[(len(A)//3*2):]
    return B,C,D
```

# Xử lí file GANNHAN

```python
gannhan = pd.read_excel("/content/drive/MyDrive/Đồ án DS201 - Deep Learning/[1] Data/GANNHAN.xlsx")
gannhan = gannhan[["ID", "path", "label"]]
gannhan = gannhan.sort_values("label")
gannhan.value_counts("path").sum()
```

```python
# Gán lại index
gannhan["index"] = range(0, len(gannhan))
gannhan = gannhan.set_index("index")
```

```python
# Thêm cột label_name
for i in range(len(gannhan)):
  if gannhan.loc[i, "label"] == 0:
    gannhan.loc[i, "label_name"] = "Normal"
  elif gannhan.loc[i, "label"] == 1:
    gannhan.loc[i, "label_name"] = "Fight"
  elif gannhan.loc[i, "label"] == 2:
    gannhan.loc[i, "label_name"] = "Fall"
  elif gannhan.loc[i, "label"] == 3:
    gannhan.loc[i, "label_name"] = "Chase"
  elif gannhan.loc[i, "label"] == 4:
    gannhan.loc[i, "label_name"] = "Bring" 
  elif gannhan.loc[i, "label"] == 5:
    gannhan.loc[i, "label_name"] = "OutOfSeat"
  elif gannhan.loc[i, "label"] == 6:
    gannhan.loc[i, "label_name"] = "NonSleep"
  elif gannhan.loc[i, "label"] == 7:
    gannhan.loc[i, "label_name"] = "Other"
  elif gannhan.loc[i, "label"] == 8:
    gannhan.loc[i, "label_name"] = "NotSuitable"  
```

```python
gannhan.sample(10)
```

```python
gannhan.value_counts("path").sum()
```

```python
# Chia các video về các nhãn tương ứng
video_per_label = {"Normal": [],
              "Abnormal": []}

for i in range(len(gannhan)):
  if (gannhan.loc[i, "label"] == 0) | (gannhan.loc[i, "label"] == 6):
    video_per_label["Normal"].append(gannhan.loc[i, "path"])
  else:
    video_per_label["Abnormal"].append(gannhan.loc[i, "path"])


# Nhãn 0 chỉ lấy 2200 video và trộn các nhãn còn lại
video_per_label["Abnormal"] = random.sample(video_per_label["Abnormal"], len(video_per_label["Abnormal"]))
video_per_label["Normal"] = random.sample(video_per_label["Normal"], 2200)
```

# Chia làm các bộ dataset

```python
# Chia data thành các tập train val test với tỉ lệ 8:1:1 theo từng nhãn
from sklearn.model_selection import train_test_split
video_train = {"Normal": [],
              "Abnormal" : []}
video_val = {"Normal": [],
              "Abnormal": []}
video_test = {"Normal": [],
              "Abnormal": []}
for label in video_per_label.keys():
  train, val_test =  train_test_split(video_per_label[label], train_size=0.8)
  val, test = train_test_split(val_test, train_size=0.5)
  video_train[label] = train
  video_val[label] = val
  video_test[label] = test
```
